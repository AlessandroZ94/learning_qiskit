```python
from qiskit.quantum_info import Statevector
from numpy import sqrt

c_0 = (1 + 2.0j) / 3
c_1 = -2 / 3
v = Statevector([c_0,c_1])
v.draw("latex")
```

```python
v.draw('bloch')
```

```python
from qiskit.visualization import plot_histogram

# The probability to find the system in state 0
expected_prob0 = abs(c_0)**2
expected_prob1 = abs(c_1)**2
print(expected_prob0)
print(expected_prob1)

statistics = v.sample_counts(1000)

display(statistics)
plot_histogram(statistics)
```
