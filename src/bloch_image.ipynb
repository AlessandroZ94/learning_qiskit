```python
from qiskit.quantum_info import Statevector, Operator
from numpy import sqrt, exp, pi

# The following image represent a qubit which is in the so called |+> state
# an equal superposition of |0> and |1>. The qubit state is displayed with an arrow
# called Bloch vector which is a vector in the unitary radius Bloch sphere
v = Statevector([1/sqrt(2),1/sqrt(2)])
v.draw("bloch")

```

```python
# Operators act on the state of qubits. The following operator
# is also called the NOT or X operator
P = Operator([[1, 0],[0, -1]])
v = v.evolve(P)

v.draw("bloch")
```
